This notebook contains plot displaying the value count of each label for the CIFAR10, CIFAR100 and FEMNIST datasets.

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
ds = load_dataset("cifar10", split="train")
unique_labels = ds.unique("label")
labels = pd.Series(ds["label"])
labels.value_counts().sort_index().to_frame().plot(kind="bar", rot=0)
print(labels.value_counts().sort_index())
plt.xlabel("Labels")
plt.title("CIFAR10 label counts")
```

```python
ds = load_dataset("cifar100", split="train")
labels = pd.Series(ds["fine_label"])
labels.value_counts().sort_index().to_frame().plot(kind="bar", rot=0)
plt.xticks(plt.xticks()[0][::5], plt.xticks()[1][::5])
print(labels.value_counts().sort_index()[:5])
plt.xlabel("Labels")
plt.title("CIFAR100 label counts")
```

```python
ds = load_dataset("adamnarozniak/femnist", split="train")
labels = pd.Series(ds["character"])
labels.value_counts().sort_index().to_frame().plot(kind="bar", rot=0)
plt.xticks(plt.xticks()[0][::3], [ds.features["character"].int2str(x) for x in labels.value_counts().sort_index().index][::3])
plt.xlabel("Labels")
plt.title("FEMNIST label counts")
```
